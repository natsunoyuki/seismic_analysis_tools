```python
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd

import os

from scipy.fft import fft, ifft
from scipy.signal import detrend

from sac_tools.Sac import Sac
```

```python
def FFT(y, t):
    dt = t[2] - t[1]
    Fs = 1.0 / dt
    L = len(y)
    Y = fft(y, L) * dt
    amp = abs(Y) 
    T = L * dt 
    freq = np.arange(0, Fs / 2, 1 / T) 
    if len(freq) < len(amp):
        amp = amp[0:len(freq)] 
    elif len(amp) < len(freq):
        freq = freq[0:len(amp)]
    return freq, amp
```
